```python

# Group data by 'theta'
theta_gpr = df.groupby('theta')
lo = theta_gpr.get_group(-1.5708).shape[0]
theta_values = df['theta'].unique()

plt.figure(figsize=(10, 6))

for theta_ in theta_values:
    data = {key: [] for key in ['x', 'y', 'z', 'theta']}
    df_ = theta_gpr.get_group(theta_)
    for i in range(lo):
        data['x'].append(df_.iloc[i]['x'])
        data['y'].append(df_.iloc[i]['y'])
        data['z'].append(df_.iloc[i]['z'])
        data['theta'].append(df_.iloc[i]['theta'])
    
    df_theta = pd.DataFrame(data)
    plt.scatter(df_theta['x'], df_theta['z'], marker=".")

dist = 35 * 1000
plt.xlim((-dist, dist))
plt.ylim((0, dist))
plt.title("Altitude vs Distance")
plt.xlabel("Distance (meters)")
plt.ylabel("Altitude (meters)")
plt.grid()
# plt.legend()
# plt.savefig("ideal_conditions.png")
plt.show()
```
