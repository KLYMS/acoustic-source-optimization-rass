```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_0 = pd.read_csv('/home/murali/Documents/rass/data/sim_data/dived_data/cpp_data/alglib_CS_0/backscatter_coordinates_filt.csv')
df_4 = pd.read_csv('/home/murali/Documents/rass/data/sim_data/dived_data/cpp_data/alglib_CS_4/backscatter_coordinates_4_filt.csv')
```

```python
print("centre coord Stats")
print(df_0.describe())
print("\nupper left coord Stats")
print(df_4.describe())
```

```python
df_04 = pd.concat([df_0,df_4])
df_04
```

```python
df_04.describe()
```

```python
plt.figure(figsize=(21, 7))

plt.subplot(1, 2, 1)
plt.scatter(df_04['x_val'],df_04['z_val'], s=0.5)

dist = 35 * 1000
plt.xlim((-dist,dist))
plt.ylim((0,dist))
plt.title('N-S')
plt.xlabel("distance")
plt.ylabel("altitude")
plt.grid()

plt.subplot(1, 2, 2)
plt.scatter(df_04['y_val'],df_04['z_val'], s=0.5)

dist = 35 * 1000
plt.xlim((-dist,dist))
plt.ylim((0,dist))
plt.title('E-W')
plt.xlabel("distance")
plt.ylabel("altitude")

plt.grid()



plt.show()
```
